```python
# openpyxl ライブラリをインストールしておくこと (pip install openpyxl / conda install openppyxl 等)
import pandas as pd

# エクセル形式のデータファイル SSDSE-B-2022 をデータフレーム ssdse_b_2022_df に読み込み、先頭5行を表示
ssdse_b_2022_df = pd.read_excel('SSDSE-B-2022.xlsx')
ssdse_b_2022_df.head()
```

```python
# 問1 SSDSE-B-2022 において、2019年の総人口が全国5番目の都道府県は？

# データフレーム ssdse_b_2022_df から年(SSDSE-B-2022の列)が2019の行を抽出し、
# 都道府県名の列(Prefecture)と総人口の列(A1101)からなるデータフレーム q1_df を作成
q1_df = ssdse_b_2022_df[ssdse_b_2022_df['SSDSE-B-2022']==2019][['Prefecture', 'A1101']]

# データフレーム q1_df の sort_valuesにより、総人口の列(A1101)で降順にソートし、上位10件を表示
print(q1_df.sort_values('A1101', ascending=False)[['Prefecture', 'A1101']].head(10))
```

```python
# 問2 2019年の日本の総人口は？

# データフレーム ssdse_b_2022_df を年の列(SSDSE-B-2022)でグルーピングして値を合計して
# 総人口の列(A1101)を抽出したデータフレーム q2_df を作成し、print文で表示
q2_df = ssdse_b_2022_df.groupby('SSDSE-B-2022').sum()['A1101']
print(q2_df)
```

```python
# 問3 2019年の(転入者-転出者)が最大の都道府県は？

# データフレーム ssdse_b_2022_df から年(SSDSE-B-2022の列)が2019の行を抽出し、
# 都道府県名(Prefecture)、転入者(A5101)、転出者(A5102)を抽出してデータフレームq3_dfを作成
q3_df = ssdse_b_2022_df[ssdse_b_2022_df['SSDSE-B-2022']==2019][['Prefecture', 'A5101', 'A5102']]

# 転入者と転出者の差をデータフレーム q3_df の diff列に追加
q3_df['diff'] = q3_df['A5101'] - q3_df['A5102']

# データフレーム q3_df の差の列(diff)を降順にソートして、上位5県を表示
print(q3_df.sort_values('diff', ascending=True).head())
```

```python
# データフレーム ssdse_b_2022_df を都道府県名(Prefecture)でグルーピングしてデータフレーム q4_df を作成
q4_df = ssdse_b_2022_df.groupby('Prefecture')

# 降水量の列(B4109)を合計した結果をデータフレーム q4_sum_df に登録
q4_sum_df = q4_df['B4109'].sum()

# データフレーム q4_sum_df を降順にソートし、上位5件と下位5件を表示
print(q4_sum_df.sort_values(ascending=False).head())
print(q4_sum_df.sort_values(ascending=False).tail())
```
