```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import japanize_matplotlib

# excelとして出力する場合
df =  pd.read_excel('ds08_apartment.xlsx')
display(df.head())
```

```python
# データのペアプロットを表示
sns.pairplot(df, height=1.5, aspect=1.0)
```

```python
# 相関行列を作成して、ヒートマップを表示
df_corr = df.loc[:, ['家賃(万円)', '専有面積(m2)', '部屋数', '駅徒歩(分)', '築年数(年)', '畳数(畳)']].corr()
sns.heatmap(df_corr, square=True, annot=True)
```

```python
# 独立変数間の分散拡大要因を確認
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

# モデルを推定し、結果を出力
X = df[['専有面積(m2)', '部屋数', '駅徒歩(分)', '築年数(年)', '畳数(畳)']]
X = sm.add_constant(X)
y = df['家賃(万円)']
# 定数項の追加
model = sm.OLS(y, X)
result = model.fit()
num_cols = model.exog.shape[1] # 説明変数の列数
vifs = [vif(model.exog, i) for i in range(0, num_cols)]
print(pd.DataFrame(vifs, index=model.exog_names, columns=['VIF']))
print(result.summary())
```

```python
# VIFが高い畳数を除いて再度モデルを推定
X = df[['専有面積(m2)', '部屋数', '駅徒歩(分)', '築年数(年)']]
X = sm.add_constant(X)
y = df['家賃(万円)']
# 定数項の追加
model = sm.OLS(y, X)
result = model.fit()
num_cols = model.exog.shape[1] # 説明変数の列数
vifs = [vif(model.exog, i) for i in range(0, num_cols)]
print(pd.DataFrame(vifs, index=model.exog_names, columns=['VIF']))
print(result.summary())
```
