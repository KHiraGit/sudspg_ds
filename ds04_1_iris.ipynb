```python
# 利用するライブラリをインポート
import pandas as pd

# irisデータを pandas のデータフレームに読み込む (ヘッダがないので、下記のように指定)
iris_df = pd.read_csv('iris.csv')
# irisデータの先頭5行分を表示
display(iris_df.head(5))
```

```python
print('## iris data (all)')
display(iris_df.describe())  # データフレーム中のデータの要約統計量を表示
```

```python
# ペアプロットを表示 (hue の指定により species により色分け、diag_kind で対角線成分にヒストグラムを表示)
import seaborn as sns
sns.pairplot(iris_df, hue='species', diag_kind='hist')
```
