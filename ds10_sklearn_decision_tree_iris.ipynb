```python
# ライブラリをインポート
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# irisデータ(csv形式)をデータテーブルの読み込み、散布図行列を表示
iris_df = pd.read_csv('iris.csv')
iris_data = iris_df[['sepal length (cm)', 'sepal width (cm)', 
                     'petal length (cm)', 'petal width (cm)']]
iris_target = iris_df['species']
sns.pairplot(iris_df, hue='species', height=2)
```

```python
clf = DecisionTreeClassifier(criterion='entropy', max_depth=5)
clf = clf.fit(iris_data, iris_target)
plt.figure(figsize=(14, 10))
plot_tree(clf, filled=True, class_names=['setosa', 'versicolor', 'virginica'],
          feature_names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],)
plt.show()
```
